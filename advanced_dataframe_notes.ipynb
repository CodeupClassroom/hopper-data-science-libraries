## And now...  SQL and Python/Pandas together at last!
- You might need to install pymysql with `python -m pip install pymysql`    
- Ensure, 100%, that you have a `.gitignore` file and it lists `env.py`
- Then we'll make our `env.py` file containing 3 plain ol' strings stored to variables:
    - `host = "ip.of.sql.server"`
    - `user = "your_username"`
    - `password = "your_password"`
- Then we'll import our values from `env.py`
- Create a connection string and store to a `url` variable
- Create a SQL query in a plain ol' string like `"select * from employees"`
- Use `pd.read_sql(query, url)` to have pandas run the SQL and return a `df`

```python
# Let's make a function that takes in a database name as a string
# this function also performs our imports from env
def get_db_url(db_name):
    from env import user, host, password
    return f'mysql+pymysql://{user}:{password}@{host}/{db_name}'
```

```python
sql = """
    SELECT * from employees limit 3000
"""
```

```python
# Called the "Connection string" b/c it has all the info to connect
url = get_db_url("employees")

employees = pd.read_sql(sql, url)
employees.head()
```

```python
# Using the employees db
# what is the most recent date in the to_date column (from titles)
url = get_db_url("employees")
sql = """
SELECT * FROM titles where to_date != '9999-01-01'
"""
titles = pd.read_sql(sql, url)
titles.head()
```

```python
titles.dtypes
```

```python
# If we want to solve the most recent date problem in pandas
titles.from_date = pd.to_datetime(titles.from_date)
titles.to_date = pd.to_datetime(titles.to_date)
titles.dtypes
```

```python
titles.head()
```

```python
titles.to_date.max()
```

```python
titles["duration"] = titles.to_date - titles.from_date
titles
```

```python
# Gossip time! Who had the shortest title duration
titles.sort_values(by="duration").head(20)
```

```python
# Could we solve this problem using only SQL?
# what is the most recent date in the to_date column (from titles)
sql = """
SELECT max(to_date) FROM titles where to_date != '9999-01-01'
"""
max_title_to_date = pd.read_sql(sql, url)
max_title_to_date.head()
```

```python
n = 40

orders = pd.DataFrame({
    'drink': np.random.choice(['Tea', 'Water', 'Water'], n),
    'meal': np.random.choice(['Curry', 'Yakisoba Noodle', 'Pad Thai'], n),
})

orders.sample(10)
```

```python
# Create a dictionary of prices for drinks and meals.

prices = {
    'Yakisoba Noodle': 9,
    'Curry': 11,
    'Pad Thai': 10,
    'Tea': 2,
    'Water': 0,
}
```

```python
orders.head()
```

```python
orders["bill"] = orders.meal.map(prices) + orders.drink.map(prices)
orders.head()
```

```python
orders.pivot_table(index="meal", columns="drink", values="bill")
```

```python
pd.crosstab(orders.meal, orders.drink, normalize=True, margins=True)
```
