```python
import matplotlib.pyplot as plt
import math
```

1. Use matplotlib to plot the following equation:

$y = x^2 -x +2$  

You'll need to write the code that generates the x and y points.

Add an anotation for the point 0, 0, the origin.

```python
x = range(-100, 100)
y = [(x**2 - x + 2) for x in x]

# Plot x and y
plt.plot(x,y)

# add title
plt.title('$ y = x^2 - x +2$')

# add labels
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.annotate('Origin', xy = (0,0), xytext = (0, 3000), arrowprops ={})
%config InlineBackend.figure_format = 'png'
```

You can use functions from the math module to help implement some of the equations above.

```python
x = range(0,100)
y = [math.sqrt(n) for n in x]

plt.plot(x,y)


plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--');
```

```python
x = range(0,20)
y = [n**3 for n in x]

plt.plot(x,y)


plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');
```

```python
x = range(0,15)
y = [2 ** n for n in x]

plt.plot(x,y)


plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.grid(True, ls = '--');
```

```python
#specify range of x values
x = range(0,15)


y = [1/(n+1) for n in x]


plt.figure(figsize = (10,6))

# plot the data
plt.plot(x,y)


# Customize the plot
# Add title
plt.title('$ y = 1/(x+1)$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')


plt.grid(True, ls = '--')

plt.show()
```

Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
x = list(range(0, 25))

# generate y for each fucntion
y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

plt.plot(x, y1, c='#30a2da', label='$\sqrt{x}$', alpha=1.0)
plt.plot(x, y2, c='orange', label='$x^{3}$', alpha=.8)
plt.plot(x, y3, c= (1,0,1), label='$2^{x}$', alpha=.8)
plt.plot(x, y4, c= '#444444', label='$1/(x+1)$', alpha=.8)


plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('My Functions')
plt.legend()

plt.show()
```
